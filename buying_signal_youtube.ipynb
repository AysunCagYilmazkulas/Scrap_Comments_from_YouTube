```python
import csv
import pandas as pd
import regex
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from utilities.buying_signal_utilities import buying_signal_utilities
```

```python
lemmatizer = WordNetLemmatizer()
buying_signal_ut=buying_signal_utilities()
signals_keyword_list = buying_signal_ut.signals_keywords()
```

```python
source ="youtube" ####### social media source
```

```python
f = open('buying_signal_youtube_unitedstates.csv', 'w')
writer = csv.writer(f)
try:
    writer.writerow(['Keyword', 'Url', 'Content', 'Comment', 'Authors_name', 'Comment_date',"Country", 'Buying_signal', 'Source'])
except Exception as e:
    print(e)

df1 = pd.read_csv("raw_files/raw_file_US/rawfile_youtube_unitedstates.csv")
df = buying_signal_ut.remove_duplicates(df1,"Comment") ############### drop repeated comments

for Keyword, Url, Content, Comment, Authors_name, Comment_date,Country, Source in df.values:

    cleaned_comment = buying_signal_ut.cleaning_text(Comment)

    # only comment from the raw file has to be cleaned


    refined_key, refined_data =buying_signal_ut.matches(Keyword, cleaned_comment)

    if refined_key == "no signal" and refined_data == "no signal":

        try:
            body = [Keyword.lower(), Url, Content, cleaned_comment, Authors_name.lower(), Comment_date, Country,'no', source]
            writer.writerow(body)

        except Exception as e:
            print(e)

    else:
        try:
            print("yes signal")
            body = [Keyword.lower(), Url, Content, cleaned_comment, Authors_name.lower(), Comment_date, Country,'yes', source]
            writer.writerow(body)
        except Exception as e:
                print(e)
f.close()
```
